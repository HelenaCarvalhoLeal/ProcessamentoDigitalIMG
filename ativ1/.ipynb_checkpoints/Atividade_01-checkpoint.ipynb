```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import matplotlib.image as mpimg
```

A figura abaixo ilustra uma família de funções radiométricas discutidas no livro texto. Escolha duas
delas para valores complementares do parâmetro γ, isto é, γ2 = 1/γ1, e aplique-as sobre a imagem
HI00008918.png , apresentada na Aula03. Além destas funções polinomiais, aplique também uma
função exponencial e sua inversa, o logaritmo.

```python
img = mpimg.imread('q1.png')

plt.imshow(img)
plt.axis('off')  # Para desativar os eixos
plt.show()
```

```python
img = io.imread('HI00008918.png')
type(img)
```

```python
# Escolhendo y1 = 5 e y2 = 0.2
def transf_radio(y):
    img_ = img / 255 # normalizando para o intervalo de 0 a 1
    img_ = np.power(img_ , [y]) # elevando os valores ao y passado como argumento
    img_ = img_ * 255 # trazendo de volta para a escala original
    img_ = img_.astype('uint8') # Deixando tudo como inteiros de 8 bits
    return img_

img_y1 = transf_radio(5)
img_y2 = transf_radio(0.2)
```

```python
# Plotando as imagens
plt.imshow(img_y1, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img_y2, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Aplicando o logaritmo natural
img_log = np.log(img)
img_log = 255 * img_log / img_log.max() 
img_log = img_log.astype('uint8')
type(img_log)
```

```python
plt.imshow(img_log, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Aplicando e
img_exp = img / 255 # Colocando no intervalo de 0 a 1
img_exp = img_exp * np.log(img.max()) # escala original utilizando o log
img_exp = np.exp(img_exp) # aplicando a e
img_exp = img_exp.astype('uint8') # colocando como inteiro de 8 bits
type(img_exp)
```

```python
plt.imshow(img_exp, cmap='gray', vmin=0, vmax=255)
plt.show()
```
