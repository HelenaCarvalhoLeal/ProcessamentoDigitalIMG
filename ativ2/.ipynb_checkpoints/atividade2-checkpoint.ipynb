```python
import numpy as np 
from skimage import io
import matplotlib.pyplot as plt
```

```python
# Criar uma imagem com ruído "uniforme"
img_uniforme = np.random.uniform(0, 2048, (512, 512))

# Exibir a imagem
plt.imshow(img_uniforme, cmap='gray')
plt.show()

# Histograma da imagem
hist_uniforme = np.histogram (img_uniforme.flatten(), bins=2048, range=(0,2048))
plt.plot (hist_uniforme[0],'-b')
plt.show()
```

```python
# definicao da funcao de transferencia
# integral do histograma da imagem de entrada
linhas, colunas = img_uniforme.shape # numero de linhas e colunas da imagem de entrada
num_pixels = linhas * colunas # numero de pixels
pr = np.histogram (img_uniforme,bins=2048,range=(0,2048))[0] / num_pixels 
T = np.cumsum (pr) # soma cumulativa
plt.plot(T, 'b')
```

```python
# Suposta função inversa, utilizamos diff para desfazer a soma cumulativa
T1 = np.diff(T, prepend=0)
plt.plot(T1, 'b')
```
