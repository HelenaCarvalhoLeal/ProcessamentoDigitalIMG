2)Gere uma imagem com ruído uniforme, assegurando que seu histograma tenha a forma desejada -
reta horizontal.Para tornar o experimento mais próximo do caso ideal, que lida com imagens 
contínuas, gere uma imagem quadrada com dimensão e profundidade 11 bits. O processo é 
semelhante ao usado para gerar imagem com ruído Gaussiano, mas não é igual.

Você deve tratar com imagens com 2^11 níveis de cinza mas não poderá visualizá-las sem perda. Os
dispositivos de apresentação, como monitores, projetores e impressoras, impõem restrição para
operar apenas com imagens com 256 níveis de cinza, mas os processadores, circuitos de memória
principal e armazenamento secundário estão livres desta limitação, portanto, o cálculo do
histograma e todas as operações matemáticas podem ser realizadas sem que seja necessário
realizar nenhum tratamento especial. Apenas na visualização da imagem é que se necessita
condicionar os valores dos pixels ao intervalo [0,256).

```python
import numpy as np 
from skimage import io
import matplotlib.pyplot as plt
```

```python
# Criar uma imagem com ruído "uniforme"
img_uniforme = np.random.uniform(0, 2048, (512, 512))

# Exibir a imagem
plt.imshow(img_uniforme, cmap='gray')
plt.show()

# Histograma da imagem
hist_uniforme = np.histogram (img_uniforme.flatten(), bins=2048, range=(0,2048))
plt.plot (hist_uniforme[0],'-b')
plt.show()
```
